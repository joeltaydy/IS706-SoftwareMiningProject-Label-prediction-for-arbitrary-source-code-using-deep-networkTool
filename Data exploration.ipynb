```python
import pandas as pd
import numpy as np
import re
import ast
import matplotlib.pyplot as plt; plt.rcdefaults()
```

# Data exploration before processed

```python
generic = lambda x: ast.literal_eval(x)
conv = {'Tags': generic}
dataset = pd.read_csv('data/data_cleaned.csv', converters=conv)
```

```python
dataset = dataset.dropna()
```

```python
dataset['Body'] = dataset.apply(lambda row: row['Body'].split("\n"), axis=1)
```

```python
tag_counter = {i:0 for i in range(1,10)}
for i, row in dataset.iterrows():
    label_list = row['Tags']
    count = len(label_list)
    tag_counter[count]+=1
```

```python
tag_num = [i for i in list(tag_counter)[:5]]
tag_frequency = [i for i in list(tag_counter.values())[:5]]
```

Surprisingly, there are some with 6 tags. <br>
We will ignore those, since the number of posts with 6 tags are only 20. <br>
Quite insigniicant compared to our dataset.

```python
objects = tag_num
y_pos = np.arange(len(objects))
freq = tag_frequency

plt.bar(y_pos, freq, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('Frequency')
plt.title('Tags per post distribution')

plt.show()
```

```python
word_count_per_posts = []
word_count_per_snippets = []
for i, row in dataset.iterrows():
    body_list = row['Body']
    words_in_post = []
    for snippets in body_list:
        word_count_per_snippets.append(len(snippets.split()))
        for word in snippets.split():
            words_in_post.append(word)
    word_count_per_posts.append(len(words_in_post))
```

```python
plt.hist(word_count_per_posts, bins=10885, alpha=0.3, color = 'blue', edgecolor = 'black')

plt.gca().set(title='Zoomed Concatenated Snippet Length Distribution', ylabel='Frequency')
plt.xlim(0,200)
plt.xlabel('Zoomed Length of Snippet (Word)')
```

```python
plt.hist(word_count_per_snippets, bins=10000, alpha=0.3, color = 'blue', edgecolor = 'black')

plt.gca().set(title='Zoomed Snippet Length Distribution', ylabel='Frequency')
plt.xlim(0,200)
plt.xlabel('Zoomed Length of Snippet (Word)')
```
