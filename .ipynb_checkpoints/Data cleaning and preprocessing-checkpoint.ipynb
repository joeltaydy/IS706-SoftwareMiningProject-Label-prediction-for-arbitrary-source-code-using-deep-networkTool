```python
import pandas as pd
import numpy as np
import re
import ast
```

# Clean the data
To get the data in the format we want
1. Getting the tags in list
2. Get the code snippet only

```python
dataset = pd.read_csv('data/data_raw.csv')
```

```python
tags = dataset["Tags"].str.slice(1,-1,1)
tags = tags.str.split("><")
dataset["Tags"] = tags
```

```python
pattern = "<code>(.*?)\</code>"
dataset['Body'] = dataset.apply(lambda row: re.findall(pattern, row['Body']), axis=1)
dataset['Body'] = dataset.apply(lambda row: "\n".join(row['Body']), axis=1)
```

```python
dataset.to_csv(r'data/data_cleaned.csv', index = False, header= True)
```

# Data preprocessing
To filter the data
1. Remove the Tags that have less than 1000 frequency
2. Remove snippets with less than 10 characters

```python
generic = lambda x: ast.literal_eval(x)
conv = {'Tags': generic}
dataset = pd.read_csv('data/data_cleaned.csv', converters=conv)
```

```python
tag_count = {}
for i, row in dataset.iterrows():
    label_list = row['Tags']
    for tag in label_list:
        if tag in tag_count:
            tag_count[tag] +=1
        else:
            tag_count[tag] = 1
```

```python
sorted_dict = {k: v for k, v in sorted(tag_count.items(), key=lambda item: item[1], reverse=True)}
```

```python
tag_filtered = {key:val for key, val in sorted_dict.items() if val > 999}
```

```python
for i, row in dataset.iterrows():
    new_tag_list=[]
    for tag in row['Tags']:
        if tag in tag_filtered:
            new_tag_list.append(tag)
    dataset.set_value(i,'Tags',new_tag_list)
```

```python
dataset = dataset.dropna()
```

```python
dataset['Body'] = dataset.apply(lambda row: row['Body'].split("\n"), axis=1)
```

```python
for i, row in dataset.iterrows():
    new_body_list=[]
    for sentence in row['Body']:
        if len(sentence) >= 10:
            new_body_list.append(sentence)
    dataset.set_value(i,'Body',new_body_list)
```

```python
dataset = dataset[dataset['Tags'].map(len) > 0]
dataset = dataset[dataset['Body'].map(len) > 0]
```

```python
dataset['Body'] = dataset.apply(lambda row: "\n".join(row['Body']), axis=1)
```

```python
dataset = dataset.drop("Id", axis =1)
```
